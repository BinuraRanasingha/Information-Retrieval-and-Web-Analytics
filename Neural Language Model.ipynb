* **re - Specifies a set of strings that matches it; the functions in this module let you check if a particular string matches a given regular expression.**


* **keras.models.Sequential - This model is appropriate for a plain stack of layers where each layer has exactly one input tensor and one output tensor. A sequential model is not appropriate when:**
1. Model has multiple inputs or multiple outputs.
2. Layers has multiple inputs or multiple outputs.
3. When we need to do layer sharing.
4. When we need a non-linear topology like a residual connection, a multi-branch model etc.

# Preproccessing the text data

```python
#Method to lower case all the words to maintain uniformity and remove words with length less than 3
def text_cleaner(text):
    #lower case text
    newString = text.lower()
    newString = re.sub(r"'s\b"," ",newString)
    
    #removing punctuations 
    newString = re.sub("[^a-zA-Z]"," ",newString)
    long_words = []
    
    #removing short words
    for i in newString.split():
        if len(i) >= 3:
            long_words.append(i)
    return(" ".join(long_words)).strip()

#preprocessing the text
data_new = text_cleaner(data_text)
```

# Creating Sequences

30 characters are taken as context and ask the model to predict the next character.

```python
def create_seq(text):
    length = 30
    sequences = list()
    
    for i in range (length, len(text)):
        #select the sequence of tokens
        seq = text[i-length:i+1]
        
        #store
        sequences.append(seq)
    print('Total sequences: %d' % len(sequences))
    return sequences

#creating sequences
sequences = create_seq(data_new)
```

```python
sequences
```

# Encoding the sequences

Encoding the sequences would convert the sequence into numbers.

```python
#Creating a character mapping index
chars = sorted(list(set(data_new)))
mapping = dict((c,i) for i,c in enumerate(chars))

def encode_seq(seq):
    sequences = list()
    for line in seq:
        
        #integer encode lines
        encoded_seq = [mapping[char] for char in line]
        
        #store
        sequences.append(encoded_seq)
        
    return sequences

#encode the sequences
sequences = encode_seq(sequences)
```

```python
sequences
```

# Creating training and validation set

```python
#vocabulary size
vocab = len(mapping)
sequences = np.array(sequences)

#creating x and y
x,y = sequences[:,:-1],sequences[:,-1]

#one hot encode y
y = to_categorical(y,num_classes=vocab)

#create train and validation sets
x_train,x_val,y_train,y_test = train_test_split(x,y,test_size=0.1,random_state=42)

print('Train shape:',x_train.shape,'Val shape:',x_val.shape)
```

# Model building

* **input_length parameter - Length of input sequences, when it is constant. This argument is required if you are going to connect Flatten then Dense layers upstream (without it, the shape of the dense outputs cannot be computed).**


* **recurrent_dropout parameter - Float between 0 and 1. It is the fraction of the units to drop for the linear transformation of the recurrent state. Default is 0.**


* **dropout parameter - Float between 0 and 1. It is the fraction of the units to drop for the linear transformation of inputs. Default 0.** 

```python
model = Sequential()
model.add(Embedding(vocab,50,input_length=30,trainable=True)) # 50 is the dimension of embedding for each character.
model.add(GRU(150,recurrent_dropout=0.1,dropout=0.1)) #150 is the dimensionality of the output space
model.add(Dense(vocab,activation='softmax'))
model.summary()
```

```python
#Compling the model:
model.compile(loss='categorical_crossentropy',metrics=['acc'],optimizer='adam')
```

* **verbose parameter - The choice of how we want to see the output of the neural network while it's training.**
1. **0 = silent** 
2. **1 = progress bar**
3. **2 = one line per epoch**

* **validation_data parameter - Data on which to evaluate the loss and any model metrics at the end of each epoch.**

```python
#Fitting the model
with tf.device('/GPU:0'):
    model.fit(x_train,y_train,epochs=100,verbose=2,validation_data=(x_val,y_test))
```

# Inference

After the model has finished training, we can generate a text from the model given an input sequence 

```python
#Method used to generate a sequence of characters with a language model
def generate_seq(model,mapping,seq_length,seed_text,n_chars):
    in_text = seed_text
    
    #generate a fixed number of characters
    for _ in range(n_chars):
        #encode the characters as integers
        encoded = [mapping[char] for char in in_text]
        
        #truncate sequences for a fixed length
        encoded = pad_sequences([encoded],maxlen=seq_length,truncating='pre')
        
        #predicting character
        yhat = model.predict_classes(encoded,verbose=0)
        
        #reverse map integer to character
        out_char = ''
        
        for char,index in mapping.items():
            if index == yhat:
                out_char = char
                break
                
        #append to input
        in_text += char
        
    return in_text
```
